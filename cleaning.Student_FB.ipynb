```python
dfs = list()

#print statement
for index, filename in enumerate(FILENAMES):
    dfs.append(pd.read_csv(filename))

dfs
```

```python
col_true_names= {'How well are the tutorials paced?':"paceing",
 'How well is the schedule paced?':"paceing",
 'How would you rate your overall satisfaction with the Summer Academy this week?':'rating',
 'Location':'location',
 'Timestamp':'timestamp',
 'Unnamed: 0':'timestamp',
 'Week':'week',
 'What track are you in?':'track'}
```

```python
# For each file:
#     Rename the headers to the defined dictionary
#    add loc and week
new_ds_set= []


for index, df in enumerate(dfs):
#       Rename the header to reflect dictionary
    rn= df.rename(columns=col_true_names)
    new_ds_set.append(rn)  
#       input new colume with filename data
    fileparts = FILENAMES[index].split("_")
    week_num, location = int(fileparts[1]), fileparts[2].split("-")[1][:-4]
#     Declare new 
    df["Location"] =  location
    df["Week"] = week_num
```
